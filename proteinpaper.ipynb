**Abstract:**

Python is an object-oriented coding language that has a wide variety of uses. Most notably for our purposes, Python can be used as a tool for data sciences and analytics.<sup>[1]</sup>
Consequently, we seek to determine the efficacy of Python as a tool to analyze the genome of viruses to identify amino-acid encoding ORFs of significance. 
For this study, we have written code to achieve this goal and analyzed the isolate Wuhan-Hu-1 strain of the SARS-CoV-2 virus as our case study.

**Methodology:**

In this study, we will define ORFs as strands of viral DNA beginning with the codon "ATG" and ending with the codon "TAG", "TAA", or "TGA". For simplicity, the known possible start codons "GTG" and "TTG" were ignored.<sup>[2]</sup> Our code analyzes viral DNA for these ORFs, including complementary strands and versions of both strands frame-shifted by one and two nucleotides. It then filters out insignificant ORFs and determines which proteins the significant ORFs encode. It is important to designate the threshold of significance for ORFs we analyze. To do this, we will use a function called ```noncoding_orf_threshold``` that shuffles the viral DNA to be analyzed a set number of times and finds the longest ORF every trial. It then will find the length of the shortest ORF found to use as the minimum cutoff threshold for significance. For this study, we used a trial size of 1,500. See below for a demonstration of the function used. 

````python
import helpers
from gene_finder import noncoding_orf_threshold

print("Please enter a valid strand of DNA containing a valid ORF.")
user_strand = input()
print("Please enter a positive integer representing the number of trials to hold.")
user_trials = int(input())
noncoding_orf_threshold(user_strand, user_trials)
````

**Conclusions:**

Our research clearly shows the capability of Python functions to identify key protein encoding ORFs in viral genomes. This has the implication that genome mapping is affordable and accessible for individuals and organizations of all types. For instance, similar methods as seen here could be used in the research of developing nations without access to vast sums of funding. 

As mentioned in the methodology, one limitation of the code we wrote was the fact it ignored the potential start codons "GTG" and "TTG". Additionally, not all notable protein encoding ORFs are extremely long, which we used as the threshold for significance. The above limitations mean that this code has the potential to miss notable ORFs, which can be an issue in the mapping of unknown genomes and new genomes. Additionally, our code has memory limitations meaning that while it is effective at analyzing the genome of a microorganism, it is not as effective at analyzing the genome of more complex organisms, for instance insects. However, both of these limitations can be addressed with further development. The first two limitations can be addressed by improving the genome analysis code to be more precise in identifying notable ORFs. The third limitation can be addressed by refactoring the genome analysis code as well as dedicating more memory and/or better hardware to running it.
